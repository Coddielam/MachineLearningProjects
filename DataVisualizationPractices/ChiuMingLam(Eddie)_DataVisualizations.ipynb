# Data Visualizations -- Line Chart and Pie Chart

```python
import pandas as pd
data = {'Morning (5:30 - 8:30 a.m.)':[0.71,0.59,0.49,0.40],
        'Evening (5:00 - 8:30 p.m.)':[0.29,0.41,0.51,0.60]} 
df = pd.DataFrame(data, index=[2016,2017,2018,2019])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))

plt.plot(df.iloc[:,0], linewidth = 3, marker = 'p', markersize=13, color='darkgrey')
plt.plot(df.iloc[:,1], linewidth = 3, marker = 'p', markersize=13, color='teal')
plt.xticks(df.index, fontsize=13)
plt.xlabel('Survey Year', fontdict={'size':15})
plt.yticks([0.3,0.4,0.5,0.6,0.7],['30%','40%','50%','60%','70%'], fontsize=13)
plt.ylabel('% of Respondents', fontdict={'size':15})
space = ' '*15
title = 'Exercise Class Time Preference Shows'+space+'\nMore Members Prefer Evening Classes'+space+'\nin Recent Years'+space+space+space+'      \n'
plt.title(title, fontdict={'size':25})
plt.legend(df.columns)
plt.tight_layout()
# plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
```

```python
data = {'#':[66,185,140,13],
        '%':[16,46,35,3]}
df = pd.DataFrame(data, index=[' 6 times','4-5 times', '2-3 times', ' Once a week'])
df
```

```python
df = df.sort_values(by='%',ascending=False)
plt.rcParams['font.size'] = 20
plt.figure(figsize=(10,7))
import seaborn as sns
cmap = sns.color_palette("Blues_r",4)
plt.pie(df.iloc[:,1], labels=df.index, autopct='%1.0f%%', explode=[0.1,0,0,0], shadow=True, colors=cmap)
space = ' '*22
title = 'Near Half of Surveyed Peak Fitness Members\nWork Out 4 to 5 Times A Week' + space + '\n'
plt.title(title)
plt.tight_layout();
```
